```
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
sns.set()
```

```
from numpy import genfromtxt

data = 'https://s3-eu-west-1.amazonaws.com/static.oc-static.com/prod/courses/files/Parcours_data_scientist/decouvrez-les-librairies-python-pour-la-data-science/hubble_data.csv'
hubble = genfromtxt(data, delimiter=',',usecols=np.arange(0,2), skip_header=1)
hubble
```

```
hubble.shape
```

```
x = hubble[::2].reshape(-1,1)
y = hubble[1::2].reshape(-1,1)
lm = LinearRegression()
model = lm.fit(x,y)
model
```

```
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
```

```
 BIGGER_SIZE = 15
plt.rc('font', size=BIGGER_SIZE) # controls default text sizes 
plt.rc('axes', titlesize=BIGGER_SIZE) # fontsize of the axes title 
plt.rc('axes', labelsize=BIGGER_SIZE) # fontsize of the x and y labels 
plt.rc('xtick', labelsize=BIGGER_SIZE) # fontsize of the tick labels
1
plt.rc('ytick', labelsize=BIGGER_SIZE) # fontsize of the tick labels 
plt.rc('legend', fontsize=BIGGER_SIZE) # legend fontsize 
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title

plt.figure() 
plt.title("Data from Hubble") 
plt.scatter(x, y, color='black', label="data") 
plt.plot(x, lm.predict(x), color='red', linewidth=3, label="fitted line") 
plt.xlabel("Distance") 
plt.ylabel("Recession velocity") 
plt.legend() 
plt.show()
```
